## Creating 3'UTR Abundance Data and Mature Abundance Data

Here were the steps taken to create 3'UTR Abundance Data and Mature Abundance Data spreadsheets

#### 3'UTR Abundance Data
To create the 3'UTR Abundance Data, I first flatten the 7mer map

```python
import pandas as pd
import numpy as np 
from numpy import loadtxt
```

```python
df = pd.read_csv("AGTC-7seq.csv", header = None)
df =pd.DataFrame(np.array(df).flatten())
df.to_csv('AGTC-7seq_flatten.csv', index= False, header = False)
```

Do the same with the frequency data (from the zebrafish 3'UTR data)

```python
df = pd.read_csv("3'UTR Frequency.csv", header = None)
df =pd.DataFrame(np.array(df).flatten())
df.to_csv("3' UTR Frequency_flatten.csv", index= False, header = False)
```

Finally, to get the normalized data, I pull the data from 3'UTR Frequency and normalize it, then print it

```python
df = pd.read_csv("3'UTR Frequency.csv", header = None)
df =pd.DataFrame(np.array(df).flatten())
df= (df - df.min()) / (df.max() - df.min()) 
df.to_csv("3' UTR Frequency_flatten_normalized.csv", index= False, header = False)
```

#### Mature Abundance Data
To create the Mature Abundance Data, I first pull that data from the dre_mature.txt

```python
data = loadtxt("dre_mature.txt", comments=">", delimiter="\z",dtype="str")
print(data)
```

Next I run the reverse complements

Finally, to get the abundance of each, I take the abundance of the sequences from the AGTC-7seq dictionary

```python
SIZE = 7
def sequence_count(string, dictionary):
    for i in range(len(string)-(SIZE-1)):
        temp_seq = string[i:i+(SIZE)].lower()
        dictionary[temp_seq]+= 1
    return dictionary

filename_data = loadtxt("reverse_dre_mature.txt", comments=">", delimiter="\z",dtype="str")

#create the dictionary
df = pd.read_csv("AGTC-7seq.csv", header = None)
data_arr = np.array(df).flatten()

data_dict = {data_arr[i]: 0 for i in range(0, len(data_arr))}

#Update the dictionary:
for i in filename_data:
    data_dict= sequence_count(i,data_dict)

for i in filename_data:
        print(data_dict[i.lower()])
```

And I do the same for the 3'UTR Sequences

```python
import csv
def get_dict(filename):
    dict_from_csv = {}

    with open(filename+'.csv', mode='r') as inp:
        reader = csv.reader(inp)
        dict_from_csv = {rows[0]:rows[1] for rows in reader}
    return dict_from_csv

data_dict = get_dict("frequency_7mer")

filename_data = loadtxt("reverse_dre_mature.txt", comments=">", delimiter="\z",dtype="str")

for i in filename_data:
        print(data_dict[i.lower()])
```
